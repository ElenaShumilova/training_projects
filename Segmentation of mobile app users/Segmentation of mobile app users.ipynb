# Выделение групп пользователей мобильного приложения на основе поведения

Перед нами данные о действиях пользователей в мобильном приложении "Ненужные вещи". В нём пользователи продают свои ненужные вещи, размещая их на доске объявлений. Наша задача, как аналитиков, — выделить группы пользователей, которые различаются по метрикам: retention rate, время, проведённое в приложении, частота действий и конверсия в целевое действие — просмотр контактов. С помощью результатов этого исследования продакт-менеджеры смогут сделать приложение ещё удобнее и популярнее.

### Преобразование типов данных

Для удобства работы переименуем столбцы и вызовем метод info() для изучения основной информации о датасетах.

```python
sourсes.rename(columns={'userId': 'user_id'}, inplace=True)
sourсes.info()
print('')
dataset.columns = dataset.columns.str.replace('.', '_')
dataset.info()
```

Данные имеют некоторые недостатки:
- не логируются сессии
- не логируется момент, когда пользователь зашёл на главный экран
- есть дубликаты событий (contacts_show и show_contacts)
- данные о времени имеют формат object вместо datetime, к тому же содержат информацию о миллисекундах

```python
dataset['event_time'] = pd.to_datetime(dataset['event_time']) #поменяем тип данных
dataset['event_time'] = dataset['event_time'].dt.round('1s') #округлим до секунд
dataset.info() #проверим
dataset.head()
```

### Объединение датафреймов по user_id

Для удобства работы объединим таблицы в одну.

```python
data = dataset.merge(sourсes, on='user_id', how='left')
data.info()
data.head()
```

### Проверка на дубликаты

```python
data['event_name'].sort_values().unique()
```

Дубликаты могут появиться из-за того, что одно и то же событие обозначено по-разному: contacts_show и show_contacts. Также у нас нет информации о различиях событий search_1 - search_7, поэтому будем считать их одним и тем же - поиском.

```python
data = data.replace({'event_name': {'show_contacts': 'contacts_show', 'search_1' : 'search', 'search_2' : 'search', 'search_3' : 'search', 'search_4' : 'search', 'search_5' : 'search', 'search_6' : 'search', 'search_7' : 'search'}})
data['event_name'].sort_values().unique()
```

```python
round((data.duplicated().sum()/len(data))*100, 2)
```

Дубликатов менее 2% в данных, поэтому их можно удалить.

```python
data = data.drop_duplicates().reset_index(drop=True)
data.duplicated().sum()
```

Теперь данные готовы к дальнейшему анализу.

### Выделение сессий

Сессии не залогированы, но мы можем добавить эти данные самостоятельно. Для этого определимся с тайм-аутом сессии. В самых распространённых системах веб-аналитики - Яндекс.Метрика и Google Analytics - сессия считается оконченной, когда в течение 30 минут пользователь не совершает новых действий (https://ru.wikipedia.org/wiki/Сессия_(веб-аналитика)). Реализуем этот подход для наших данных.

```python
#отсортируем данные по user_id и времени события
data = data.sort_values(['user_id', 'event_time']).reset_index(drop=True)
#определим разницу в 30 минут для выделения группы событий
g = (data.groupby('user_id')['event_time'].diff() > pd.Timedelta('30Min')).cumsum()
#добавим счётчик групп, который и будет идентификатором сессий
data['session_id'] = data.groupby(['user_id', g], sort=False).ngroup() + 1
data.head()
```

### Сколько в среднем длится сессия?

Найдём сначала продолжительность каждой сессии.

```python
sessions = data.groupby('session_id', as_index=False).agg({'event_time': ['min', 'max']})
sessions.set_axis(['session_id', 'session_start', 'session_finish'], axis=1, inplace=True)
sessions['session_duration'] = sessions['session_finish'] - sessions['session_start']
display(sessions.head())
sessions['session_duration'].describe()
```

У нас явно есть выбросы. Во-первых, есть сессии длиной 0 секунд. Такое могло произойти, например, из-за округления миллисекунд. Во-вторых, максимальная длина сессии больше 5 часов. Для удобства работы с выбросами переведём данные стобца `duration` в секунды.

```python
sessions['session_duration'] = sessions['session_duration'].astype('timedelta64[s]')
sessions.info()
```

```python
sns_plot = sns.distplot(sessions['session_duration'])
fig = sns_plot.get_figure()
plt.xlabel('Количество сессий')
plt.ylabel('Продолжительность сессии в секундах')
plt.title('Распределение продолжительности сессий')
plt.show()
```

```python
null = sessions.query('session_duration == 0')
round((len(null) / len(sessions)*100), 2)
```

Мы видим, что сессий продолжительностью 0 секунд много: более 20%. Возможно, есть какие-то технические неполадки в работе приложения или регистрации событий.

Посмотрим теперь на слишком долгие сесии.

```python
np.percentile(sessions['session_duration'], [90, 95, 99])
```

Больше 5538 секунд в приложении провели не более 1% пользователей. Это чуть больше 92 минут, что выглядит уже более реально, чем 5 часов. Отбросим аномально длинные сессии и те, что длились 0 секунд, они для нас неинформативны.

```python
sessions = sessions.query('(session_duration != 0) & (session_duration < 5538)') 
sessions['session_duration'].describe()
```

```python
sns_plot = sns.distplot(sessions['session_duration'])
fig = sns_plot.get_figure()
plt.xlabel('Количество сессий')
plt.ylabel('Продолжительность сессии в секундах')
plt.title('Распределение продолжительности сессий после удаления аномалий')
plt.show()
```

Удалим данные об этих сессиях из исходного датафрейма.

```python
data = data[data['session_id'].isin(sessions['session_id'].tolist())]
```

```python
data.to_csv('apps1.csv')
```

### Расчёт retention rate для каждого сегмента

Определим время первого действия для пользователя в приложении.

```python
first_ts = (data
            .sort_values(['user_id', 'event_time'])
            .reset_index(drop=True)
            .groupby('user_id', as_index=False)
            .agg({'event_time': 'min'})
            .set_axis(['user_id', 'first_ts'], axis=1)
           )
first_ts.head()
```

Зададим функции, необходимые для создания профилей пользователей каждого сегмента и для расчёта удержания.

```python
# функция для созжания профилей пользователей
def get_profiles(segment, first_ts):
    profiles = (segment[['user_id', 'session_start']]
                      .merge(first_ts, on='user_id', how='left')
                      .drop_duplicates()
                      .reset_index(drop=True)
                     )
    profiles['dt'] = profiles['first_ts'].dt.date
    return profiles
```

```python
#функция для расчёта удержания
def get_retention(
    profiles, observation_date, horizon_days, ignore_horizon=False
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw.insert(4, 'lifetime', (result_raw['session_start'] - result_raw['first_ts']).dt.days)

    # рассчитываем удержание
    result_grouped = result_raw.pivot_table(
        index=['dt'], columns='lifetime', values='user_id', aggfunc='nunique'
    )
    cohort_sizes = (
        result_raw.groupby('dt')
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    result_grouped = cohort_sizes.merge(
        result_grouped, on='dt', how='left'
    ).fillna(0)
    result_grouped = result_grouped.div(result_grouped['cohort_size'], axis=0)

    # исключаем все лайфтаймы, превышающие горизонт анализа
    result_grouped = result_grouped[
        ['cohort_size'] + list(range(horizon_days))
    ]

    # восстанавливаем столбец с размерами когорт
    result_grouped['cohort_size'] = cohort_sizes

    # возвращаем таблицу удержания и сырые данные
    return result_raw, result_grouped
```

Всё готово для расчёта таблицы удержания пользователей каждого сегмента. Возьмём горизонт анализа 7 дней.

```python
# для первого сегмента
profiles_first = get_profiles(segment_first, first_ts)
retention_raw_first, retention_first = get_retention(
    profiles_first, datetime(2019, 11, 3).date(), 7
)

retention_first
```

## Проверка гипотез

### Пользователи, установившие приложение по ссылке из yandex и из google, демонстрируют разную конверсию в просмотр контактов

Нулевая гипотеза будет звучать так: конверсия пользователей, установивших приложение по ссылке из yandex и из google, не различается. Альтернативная гипотеза: конверсия пользователей, установивших приложение по ссылке из yandex и из google, различается.

Критический уровень статистической значимости зададим в 5%, это стандарт, который нас устраивает.

```python
# Создаём список с количеством попыток
trials = (data
          .query('source !="other"') #оставляем только данные для яндекс и гугл
          .groupby('source', as_index=False)['user_id'].nunique() #считаем количество уникальных пользователей для каждого источника
          ['user_id'].tolist() #из столбца с количеством формируем список
         ) 
trials
```

```python
# Создаём список с количеством успехов
successes = (data
          .query('(source !="other") & (event_name == "contacts_show")') #оставляем только данные для яндекс и гугл
          .groupby('source', as_index=False)['user_id'].nunique() #считаем количество уникальных пользователей для каждого источника
          ['user_id'].tolist()
            ) #из столбца с количеством формируем список
successes
```

```python
alpha = .05

# пропорция успехов в первой группе:
p1 = successes[0]/trials[0]

# пропорция успехов во второй группе:
p2 = successes[1]/trials[1]

# пропорция успехов в комбинированном датасете:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# разница пропорций в датасетах
difference = p1 - p2 

# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)  

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    ) 
```

- Самое частое событие - это показ рекомендованных объявлений. Также пользователи часто просматривают фото, карточки товара и пользуются функцией поиска. Пути пользователя к целевому действию могут быть весьма многообразны. Даже само целевое действие может оказаться первым. Это возможно, если пользователь перешёл по прямой ссылке. 

- Мы сегментировали пользователей на основании среднего времени сессии и получили три группы: тех, кто проводит в приложении до 11 минут, от 11 до 21 минуты и свыше 21 минуты.

- Лучше всего удерживаются пользователи 2 сегмента. По частоте действий лидируют пользователи 1 сегмента. Конверсия в целевое действие при этом выше у пользователей 3 сегмента.

- Мы сформулировали и проверили 2 статистические гипотезы. Выяснили, что конверсия пользователей, установивших приложение по ссылке из yandex и из google, не различается. А также узнали, что конверсия пользователей, совершающих и не совершающих действие tips_click (кликнуть по рекомендованному объявлению), различается.

### Рекомендации для продакт-менеджера

- Пользователи видят очень много рекомендаций, но редко по ним кликают. На это стоит обратить внимание, поскольку те, кто кликнул, лучше конвертируются. Можно посмотреть, насколько эти рекомендации удобны при навигации и релевантны. Для многих показ рекомендованного оказался последним событием в сессии, возможно, их слишком много и они кажутся назойливыми.

- Те пользователи, которые проводят много времени в приложении, лучше конвертируются, но хуже удерживаются. Возможно, для таких пользователей окажется эффективной система начисления бонусных баллов за несколько дней использования подряд.

- Малое количество действий за сессию и низкое удержание может быть связано с неудобным интерфейсом, стоит обратить на это внимание и проработать, если обнаружатся недостатки.
